```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz
%matplotlib inline
```

```python
x1 = np.linspace(-10, 10)
x2 = np.linspace(-10, 10)
membership_x1 = fuzz.trimf(x1, [-10, -3.75, 2.5])
membership_x2 = fuzz.trimf(x2, [-2.5, 3.75, 10])
sinc = np.outer(np.sin(membership_x1), np.sin(membership_x2)) / np.outer(membership_x1, membership_x2)
```

```python
plt.imshow(sinc)
plt.show()
```

```python
plt.plot(sinc)
plt.show
```

```python
plt.plot(x1, membership_x1)
plt.grid(True)
plt.show()
```

```python
plt.plot(x2, membership_x2)
plt.grid(True)
plt.show()
```

```python
print(x1[np.argmax(membership_x1)])
```

```python
print(x2[np.argmax(membership_x2)])
```

```python
plt.plot(x1, membership_x1)
plt.plot(x2, membership_x2)
plt.grid(True)
plt.show()
```

```python
sinc_no_fuzz = np.outer(np.sin(x1), np.sin(x2)) / np.outer(x1, x2)
plt.plot(sinc_no_fuzz)
plt.show
```
