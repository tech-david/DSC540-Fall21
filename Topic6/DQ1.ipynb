```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = np.arange(-1, 10, 1)
trap_points = [-1, 4, 5, 9]
y = fuzz.trapmf(X, trap_points)
plt.plot(X, y)
plt.axvline(X.item(5),
            ymin=y.min(),
            ymax=y.max())
plt.axvline(X.item(6),
            ymin=y.min(),
            ymax=y.max())
plt.xticks(np.arange(-1, 9, step=1))
plt.show()
```
